```python
import cv2
import numpy as np
```

```python
image = cv2.imread('mountain.jpg')
image
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray
```

```python
result = np.zeros((image.shape[0], 256), dtype=np.uint8)
result
```

```python
print(image.shape, gray.shape, result.shape)
```

```python
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
# (원본 이미지, 0번 채널, mask=None, 히스토그램 사이즈=256(0~255), range)
cv2.normalize(hist, hist, 0, 255, cv2.NORM_MINMAX)
```

```python
print(hist.shape)
```

```python
for x, y in enumerate(hist):
    cv2.line(result, (int(x), image.shape[0]), (int(x), image.shape[0] - int(y)), 255)
```

```python
dst = np.hstack([image[:,:,0], result])
```

```python
cv2.imshow('dst', dst) # 창 이름 설정
```
